```python
import os
import librosa as librosa
import IPython.display as ipd
```

#looking at our directories:

```python
print(os.listdir())
print(os.listdir('orchestral_samples'))
print(os.listdir('orchestral_samples/data'))
print(os.listdir('orchestral_samples/data/mandolin'))
```

Okay loading our files

```python
filename = 'orchestral_samples/data/mandolin/mandolin_B3_very-long_piano_tremolo.wav'
y, sr = librosa.load(filename)
```

```python
print(type(y), '\n sample rate: ', sr)
print('length: ',  len(y))
print('duration:', len(y)/sr, ' seconds')
```

Okay embedding the audio in a notebook

```python
audio_element = ipd.Audio(filename)
ipd.display(audio_element)
```
