```python
import librosa
import crepe
import matplotlib.pyplot as plt
import numpy as np
import IPython.display as ipd
```

```python
filename = 'orchestral_samples/data/mandolin/mandolin_B3_very-long_piano_tremolo.wav'
y, sr = librosa.load(filename)
```

```python
time, frequency, confidence, activation = crepe.predict(y, sr, viterbi=True)
```

```python
plt.plot(time,frequency)
```

```python
freqs = librosa.hz_to_note(frequency)

plt.plot(time,freqs)
```

```python
plt.plot(time,confidence)
```

```python
low = librosa.note_to_hz("e1") 
high = librosa.note_to_hz("e5")
frequ,prediction,confidence = librosa.pyin(y, low , high, sr)
```
